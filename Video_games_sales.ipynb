```python
def platform_cleaning(i):
    if i == 'GB':
        return 'Game Boy'
    elif i == 'X360':
        return 'XBox 360'
    elif i == 'GBA':
        return 'Game Boy Advanced'
    elif i == 'XB':
        return 'XBox'
    elif i == '2600':
        return 'Atari 2600'
    elif i == 'XOne':
        return 'XBox One'
    elif i == 'GC':
        return 'Game Boy Color'
    elif i == 'GEN':
        return 'Genesis'
    elif i == 'DC':
        return 'Dreamcast'
    elif i == 'PSV':
        return 'PS5'
    elif i == 'SAT':
        return 'Sega Saturn'
    elif i == 'SCD':
        return 'Sega CD'
    elif i == 'NG':
        return 'Neo Geo'
    elif i == 'GG':
        return 'Game Gear'
    else:
        return i
```

```python
def cleaning_year(df, col1, col2):
    droping_conditions = (df[col1].isna()) & (df[col2] == 'Unknown')
    df.loc[droping_conditions, col1] = 'Drop'
```
